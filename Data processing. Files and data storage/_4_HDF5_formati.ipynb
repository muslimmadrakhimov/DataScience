HDF5  formati

katta fayllarni saqlaganda siqilgan holatda saqlaydi

```python
# HDFStore klassidan obyekt yaratish
hdfobj = pd.HDFStore("data.h5") # data.h5 degan fayl yaratib oldik
```

Bizda hozir bir nechta DataFrame lar mavjud, shulardan foydalanib HDFStore klassiga saqlaymiz

```python
# df dataframe da Uzbekiston haqida malumot bor
# df1 dataframe da cars yani mashinalar haqida malumot bor
# shu ikkala DataFrame faylini yangi obyekt hdfobj tiga yani HDF5 formatida saqlaymiz
hdfobj.put("uzdata", df) # uzdata degan faylga df dagi uzbekiston haqida malumotni saqlaymiz
```

```python
# yangi cardata fayl yaratib unga df2 dagi avtomobil malumotlarini saqlaymiz
hdfobj.put("cardata", df2)
```

```python
# HDFOBJ  obyektini kalitlarini kórish
hdfobj.keys()
```
