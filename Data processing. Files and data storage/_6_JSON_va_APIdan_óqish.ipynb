JSON formatidagi faylni jadval kórinishida óqib olish usuli

```python
# requests  komandasini chaqirib olamiz
import requests
# requests ichida get metodini chqirib olamiz
r = requests.get("https://openlibrary.org/subjects/python.json?published_in=2010-2021")
```

```python
# saqlab olingan LINK ni tógri ishladimi yoqmi kórish uchun
# buning uchun (status_code) parametrini yozamiz
# komanda bajarilganda natija 200 chiqsa tógri natija chiqqan bóladi
# 200 dan boshqa qiymatlar chiqsa ishlamagan bóladi
r.status_code
```

JSON  natijani kórish

```python
# bu ósha link dagi json kórinishidagi kitob haqida ma'lumotlar
r.json()
```

JSON kórinishidagi ma'lumotlarni jadval kórinishiga yani DataFrame ga ótkazish usuli

```python
pd.DataFrame(r.json())
```
