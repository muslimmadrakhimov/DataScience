BETWEEN - ORALIQDA DEGANI

```python
# BETWEEN - oraliqda, yani 2005 va 2015 yil oraligidagi malumotlarni chiqaradi
query = "SELECT country, year, coal, gas, hydro, oil FROM elect WHERE country='Uzbekistan' AND year BETWEEN 2005 AND 2015"
pd.read_sql(query,db)
```

IN - komandasi

```python
# IN - komandasi bizga aynan qaysi qiymat kerakligini chiqaradi
query = "SELECT country, year, coal, gas, hydro, oil FROM elect WHERE country='Uzbekistan' AND year IN  (1990, 1995, 2000, 2005)"
pd.read_sql(query,db)
```

LIKE - komandasi jadavaldan qidirish uchun ishlatiladi

```python
# LIKE  - komandasi yordamida teks qiymatlarda ohirgi qatordagi harflardan qiymatlarni topsa bõladi
# %var - Anvar, Sarvar,
# _ - _ar -> car, bar, tar,
# __ar - star,
# %ol -> bola, xola, bolalar,

query = "SELECT * FROM elect WHERE country LIKE '%stan' AND year IN (2019, 2020, 2021)"
pd.read_sql(query,db)
```

```python
# A bilan boshlangan barcha qiymat sõzlarni chiqarish
query = "SELECT * FROM elect WHERE code LIKE 'A__' AND year=2020"
pd.read_sql(query,db)
```
