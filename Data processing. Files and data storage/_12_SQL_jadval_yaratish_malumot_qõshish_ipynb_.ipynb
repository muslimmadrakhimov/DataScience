Kerakli modullarni chaqirib olamiz

```python
import pandas as pd
import sqlite3 as sql
```

```python
# yangi malumotlar ombori yaratamiz
# bu malumotlar ombori yangi ichida fayl yōq
db = sql.connect("newdb.db")
cursor = db.cursor()
```

Yangi malumotlar omboriga malumotlar joylash

```python
# faylda turli davlatlarning elektrik manbalari keltirilgan
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/Electricity_Production_By_Source.csv")
df.head()
```

```python
# ustunlar uchun rõyhat yaratamiz
cols = ['country', 'code', 'year', 'coal', 'gas', 'hydro', 'renewable_other', 'solar', 'oil', 'wind', 'nuclear']
# DataFrame ustunlari nomi uzun bõlgani uchun ularni qisqartirib õzgartiramiz
df.columns = cols
df.head()
```

```python
# DataFrame kōrinishidagi faylni SQl omboriga õtkazish
df.to_sql('elect',db)
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall)
```

```python
cursor.execute("PRAGMA table_info(elect);")
cursor.fetchall()
```

```python
# bizga kerakli bõlgan malumotlarni chiqarish, masalan Uzbekistan haqida
query = "SELECT * FROM elect WHERE country='Uzbekistan'"
pd.read_sql(query,db)
```

SQL JADVAL YARATISH_MALUMOT QÕSHISH

```python
# """ - 3 ta qõshtirnoq matnlarni bir necha qatorlarga yozishga imkon beradi
# rank, country, density, population, area bu ustunlar nomlari hisoblanadi
query = """
CREATE TABLE IF NOT EXISTS worldpop(
rank INT,
country TEXT PRIMARY KEY,
density DOUBLE,
population DOUBLE NOT NULL,
area INT NOT NULL
);
"""
cursor.execute(query)
db.commit() # jadvalga ōzgartirish kiritganimiz uchun commit()-saqlash komandasini yozamiz
```

```python
# jadval qõshilganini tekshirib kõramiz
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
cursor.fetchall()
```

```python
# worldpop jadvalini tarkibini kõrish
cursor.execute("PRAGMA table_info(worldpop);")
cursor.fetchall()
```

```python
# jadvalga malumot qõshish, qiymatlar kiritish
query = "INSERT INTO worldpop VALUES (1, 'Macau', 21946.4667, 658.3940, 30)"
cursor.execute(query)
db.commit()
```

```python
# tekshirib kõramiz, yani worldpop jadvalidan hamma qiymatlarni olish
cursor.execute("SELECT * FROM worldpop")
cursor.fetchall()
```

```python
# aynan bizga kerakli bõlgan ustunlarga qiymat beramiz, masalan (country, population, area)
# VALUES - qiymatlar, ustunlarga qiymatlar beramiz
query = "INSERT INTO worldpop (country, population, area) VALUES ('Uzbekistan', 33935.7630, 447400)"
cursor.execute(query)
db.commit()
```

```python
# tekshirib kõramiz, yani worldpop jadvalidan hamma qiymatlarni olish
cursor.execute("SELECT * FROM worldpop")
cursor.fetchall()
```
