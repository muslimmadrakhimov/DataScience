Ma'lumotlar omboriga ulanish

```python
db = sql.connect('chinook.db')
# cursor metodi yordamida sql ni ishga tushirish
cursor = db.cursor()
```

```python
# Malumotlar ombori ichidagi jadvallarni kórish usuli, dastur yordamida
# execute - bajarish komandasi
# qavsni ihidagi yozuv SQL tilida yozilgan, SQL komandasi

cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
cursor.fetchall() # fetchall - barchasini kórish
# natija bir nechta jadvallar chiqadi
```

Jadavaldi ichidagi qatorlar sonini kórish

```python
cursor.execute("SELECT COUNT (*) FROM sqlite_sequence")
cursor.fetchone()
```

```python
cursor.execute("SELECT COUNT (*) FROM customers")
cursor.fetchone()
```

Jadvalni ichidagi ustunlar sonini kórish

```python
cursor.execute("PRAGMA table_info(customers);") # PRAGMA komandasi SQLITE3 hos komanda
cursor.fetchall()
```

JADVALNI DF ga ÕQISH

```python
cursor.execute("SELECT * FROM customers") # * yulduzcha barcha qiymatlar degani
mijozlar = cursor.fetchall()
# malumotlar omboridagi barcha mijozlar malumotlari chiqadi rõyhat kõrinishida
# Rõyhat kõrinishda uzun bōlgani uchun 'mijozlar' ga saqlab qõyamiz
```

DataFrame kõrinishida chiqarish

```python
df = pd.read_sql("SELECT * FROM customers", db, index_col='CustomerId')
df
```

```python
# DF da kerakli qatorlarni olish uchun LIMIT 10 yozsek
# Boshidan 10 ta qatorni chiqaradi
df = pd.read_sql("SELECT * FROM customers LIMIT 10", db, index_col='CustomerId')
df
```

```python
# Bizga kerakli bõlgan ustunlarni chiqarish usuli
df = pd.read_sql("SELECT CustomerId, Firstname, Lastname, Phone FROM customers", db, index_col="CustomerId")
df.head()
```

TARTIBLASH(ORDER BY) VA SARALASH(DISTICT)

```python
# sql komandasi uzun bõlgani uchun alohida saqlab olamiz, komanda degan õzgaruvchiga
komanda = "SELECT CustomerId, Firstname, Lastname, Phone, Country FROM customers"
df = pd.read_sql(komanda, db, index_col="CustomerId")
df.head()
```

Tartiblash

```python
# Familiyasi bõyicha tartiblash
komanda = "SELECT CustomerId, Firstname, Lastname, Phone, Country FROM customers ORDER BY LastName"
df = pd.read_sql(komanda, db, index_col="CustomerId")
df
```

```python
# Faqat davlatlar rõyhatini chiqarish usuli
# customers faylini ichidan davlatlar rõyhatini takrorlamasdan(DISTINCT) tanlagin degan komandani beramiz
query = "SELECT DISTINCT Country FROM customers"
cursor.execute(query)
cursor.fetchall()
```

```python
# DataFrame kõrinishida chiqarish usuli
pd.read_sql(query,db)
```
