.isin()  metodi

```python
# .isin() metodi DataFrame yoki Series ichidagi qiymatlarning berilgan ro'yxat yoki to'plamdagi qiymatlarga mos kelishini tekshiradi
# va boolean Series (True yoki False) qaytaradi.
models1 = [' SLK', ' B Class']
filtr = df['model'].isin(models1) # model ichidagi ' Slk', ' B Class' qiymatlarini tekshirish va bōlsa chiqarish usuli
filtr
```

```python
# saqlangan filtr obyektini DataFrame ga qōllash yani ajratib filtrlab olish
df[filtr] # bizga kerak bōlgan aynan ' Slk', ' B Class' qiymatlarni chiqardik
```

```python
# saqlab olish
df2 = df[filtr]
df2
```
