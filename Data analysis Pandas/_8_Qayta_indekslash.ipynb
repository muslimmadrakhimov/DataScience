```python
import pandas as pd
import numpy as np
```

```python
data = {'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
        'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
        'Temp':[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53],
        'Zichlik':[79.77, 78.68, 77.53, 76.34, 75.13, 73.91, 72.71, 67.03]}
df = pd.DataFrame(data)
df
```

Index larni qiymatini ōzgartirish

```python
df.index = [21, 20, 19, 18, 17, 16, 15, 10]
df
```

```python
# Reindex
ind = [21, 20, 19, 18, 17, 16, 15, 10]
# rōyhatni tartiblash
ind.sort()
ind
```

```python
df.reindex(ind)
```

```python
# DataFrame ni yani df ni chiqaradigan bōlsak df indexi ōzgarmadi
df
```

```python
# saqlash usuli
df = df.reindex(ind)
df
```

```python
# reindex orqali kerakli index larni olib chiqarish usuli
df.reindex([10, 15, 20])
```

```python
# yōq malumotga NAN yani yōq malumot kōrsatadi
df.reindex(index=range(10, 22), method='nearest') # nearest eng yaqin qiymat degani
```

```python
# ffill  metodi eng ohirgi qiymatni olib yōq qiymatlarni tōldirib chiqadi
df.reindex(index=range(10, 22), method='ffill')
```
