```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/Electricity_Production_By_Source.csv")
df.columns = ['davlat','kod','yil','komir','gaz','gidro','yangilanuvchi','quyosh','neft','shamol','atom']
df.head()
```

### 3-VAZIFA. DF dan faqat AQSh (`United States`) uchun ma'lumotlarni ajratib oling va yangi `dfus` nomli o'zgaruvchiga saqlang

```python
# JAVOBNI SHUYERGA YOZING
# AQSh (United States) uchun ma'lumotlarni ajratib olish
dfus = df[df['davlat'] == 'United States']

# Yangi dataframe'ni ko'rsatish
dfus.head(6)
```

### 4-VAZIFA. `dfus` jadvalining indeksini `yil` ustuni bilan o'zgartiring. DF dan `yil` ustunini tashlab yuboring.

```python
# JAVOBNI SHUYERGA YOZING

# 'yil' ustunini indeksga o'zgartirish
dfus = dfus.set_index('yil')

# Asl DF dan 'yil' ustunini tashlab yuborish
df = df.drop(columns=['yil'])

# Yangi dfus va df ko'rsatish
dfus.head()
```

```python
#@title Kutilgan natija
```

### 5-VAZIFA. AQSh 2020-yilda jami qancha elektr energiyasi ishlab chiqqan?

```python
# JAVOBNI SHUYERGA YOZING
# 2020-yil uchun jami energiyani hisoblash
jami_energiya_2020 = dfus.loc[2020, ['komir', 'gaz', 'gidro', 'yangilanuvchi', 'quyosh', 'neft', 'shamol', 'atom']].sum()

# Natijani ko'rsatish
jami_energiya_2020
```

```python
#@title O'xshash natija (2018-yil uchun)
```

### 6-VAZIFA. 2010-2020-yillarda AQSH turli manbaalardan o'rtacha qancha elektr toki olganini chiqaring. Natijani kichikdan kattaga qarab tartiblang.

```python
# JAVOBNI SHUYERGA YOZING

# Yil ustuni bo'yicha filtrlangan ma'lumotlar
dfus_2010_2020 = dfus.loc[2010:2020]

# Faqat raqamli ustunlarni tanlash (davlat va kod ustunlarini olib tashlash)
dfus_2010_2020 = dfus_2010_2020.drop(columns=['davlat', 'kod'])

# Har bir manba bo'yicha o'rtacha elektr toki miqdorini hisoblash
average_energy = dfus_2010_2020.mean()

# Natijani kichikdan kattaga qarab tartiblash
average_energy_sorted = average_energy.sort_values()

# Natijani ko'rish
average_energy_sorted
```

```python
#@title O'xshash natija (1990-2000 yillar uchun)
```

### 7-VAZIFA. AQSH gidro elektr stansiyalardan eng kam va eng ko'p elektr energiyasi olgan yillarni toping.

```python
# JAVOBNI SHUYERGA YOZING


# AQSH uchun gidroelektr energiyasi ma'lumotlarini olish
hydro_energy = dfus['gidro']

# Eng kam va eng ko'p elektr energiyasi olingan yillarni topish
min_year = hydro_energy.idxmin()  # eng kam energiya olingan yil
max_year = hydro_energy.idxmax()  # eng ko'p energiya olingan yil

# Natijani chiqarish
min_year, hydro_energy[min_year], max_year, hydro_energy[max_year]
```

### 9-VAZIFA. 2019-yilda har bir davlat jami qancha elektr energiya ishlab chiqarganini DF ko'rinishida chiqaring.
*ishora. `Series` obyektini `DF` ga o'tkazing


```python
# JAVOBNI SHUYERGA YOZING

# 2019-yilni ajratib olish
df19 = df[df['yil'] == 2019]

# Har bir davlat bo'yicha jami elektr energiya ishlab chiqarishni hisoblash (barcha manbaalar yig'indisi)
df19['jami_energiyasi'] = df19[['komir', 'gaz', 'gidro', 'yangilanuvchi', 'quyosh', 'neft', 'shamol', 'atom']].sum(axis=1)

# Faqat davlat va jami energiya ustunlarini olish
df19_series = df19[['jami_energiyasi']]

# Indeksni davlatga o'zgartirish
df19_series = df19_series.set_index(df19['davlat'])

# Natijani ko'rish
df19_series
```

### 10-VAZIFA.
#### 10.1. Asl DF dan 2000-2020-yillar, `France` uchun `atom`, `komir`, `gaz`, `quyosh` manbaalarini alohida `dffr` jadvaliga ko'chirib oling
#### 10.2. `dffr` jadvalini yil ustuni bo'yicha tartiblang

```python
# JAVOBNI SHUYERGA YOZING
# 2000-2020-yillar orasida faqat France uchun atom, komir, gaz, quyosh manbaalarini ajratib olish
dffr = df[(df['yil'] >= 2000) & (df['yil'] <= 2020) & (df['davlat'] == 'France')][['yil', 'atom', 'komir', 'gaz', 'quyosh']]

# Yil ustuni bo'yicha tartiblang
dffr = dffr.sort_values(by='yil')

# Natijani ko'rish
dffr
```
