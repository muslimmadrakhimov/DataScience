```python
import pandas as pd
import numpy as np
```

Series obyektlari ustida arifmetik amallar

```python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'c', 'd', 'e'])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=['a', 'c', 'e', 'f', 'g'])
```

```python
s1
```

```python
s2
```

```python
s1 + s2
```

DataFrame obyektlari ustida arifmetik amallar

```python
df1 = pd.DataFrame(np.arange(9.).reshape(3,3), columns=list('bcd'), index=['Olma', 'Anor', 'Uzum'])
df2 = pd.DataFrame(np.arange(12.).reshape(4,3), columns=list('abc'), index=['Olma', 'Anor', 'Qovun', 'Anjir'])
```

```python
df1
```

```python
df2
```

Qōshish usuli

```python
df1 + df2 # qōshish usuli
```

```python
# qōshishni yana bir usuli
df1.add(df2)
```

```python
df1.add(df2, fill_value=0) # fill_value metodi  yōq bōlgan qiymatlar ōrniga qiymat qōyib ketadi
```

Ayirish usuli

```python
df2.sub(df1, fill_value=0)
```

```python
df2.rsub(df1, fill_value=0) # rsub metodi teskarisi ayirish
```

Bōlish metodi

```python
df2.div(df1)
```

```python
df1.div(df2)
```

DataFrame va Series obyektlari ōrtasida arifmetik amallar

```python
df = pd.DataFrame(np.arange(1, 13.).reshape(3, 4), index=list('abc'), columns=['Olma', 'Anor', 'Qovun', 'Anjir'])
obj = pd.Series([1,1,2,2], index=df.columns)
```

```python
df
```

```python
obj
```

```python
# DataFrame va Series obyekylarini qōshish
df + obj
```

```python
obj = pd.Series([2, 3, 1], index=['Olma', 'Uzum', 'Qovun'])
obj
```

```python
df.sub(obj)
```

```python
obj['Olma']
```

```python
df + obj['Olma']
```
