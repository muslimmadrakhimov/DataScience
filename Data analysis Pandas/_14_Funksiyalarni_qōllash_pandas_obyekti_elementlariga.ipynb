```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(3, 4), index=list('abc'), columns=['Olma', 'Anor', 'Qovun', 'Anjir'])
df
```

abs (absolyut) metodi qiymatlar manfiy yoki musbat bōlsa ham bir hil musbat kōrinishida chiqarib beradi

```python
np.abs(df)
```

round butunlash metodi

```python
np.round(df) # hamma qiymarlarni butun son qiymatiga ōzgartiradi
```

mean()  metodi ōrtacha qiymatni chiqaradi

```python
df.mean() # ustunlari bōyicha ōrtacha qiymat
```

```python
df.mean(axis=1) # qatorlari bōyicha ōrtacha qiymat
```

.apply  metodi  yordamida qōshimcha metodlar yozish mumkin

```python
df.apply(np.abs)
```

```python
df.apply(np.sum)
```

```python
df.apply(sum, axis=1) # ustunlar bōyicha chiqarish
```

```python
# funksiya
f = lambda x: x.max() - x.min()
```

```python
df.apply(f)
```

```python
df.apply(lambda x: x.max() - x.min()) # yana bir usul yozilishi
```

```python
def foo(x):
  return pd.Series([x.min(), x.max()], index=['min', 'max'])
```

```python
df.apply(foo)
```

.applymap  metodi har bitta elementga qōllaniladi

```python
df = pd.DataFrame(np.arange(1, 13).reshape(3, 4), index=list('abc'), columns=['Olma', 'Anor', 'Qovun', 'Anjir'])
df
```

```python
# har bitta elementni kvadratini hisoblash funksiyasi
df.applymap(lambda x: x*x)
```
