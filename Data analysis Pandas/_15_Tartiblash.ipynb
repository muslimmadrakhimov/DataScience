DataFrame va Series obyekt elementlarini tartiblash (sort), reytinglash (ranking)

```python
obj = pd.Series(range(4), index=['d', 'a', 'b', 'c'])
```

```python
obj
```

```python
# qiymatlar bōyicha tartiblash
obj.sort_values()
```

```python
# KEling keyingi mashqlar uchun Githubdan kattaroq DF yuklab olamiz
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
# DataFrame hajmini kōrish uchun .shape  metodi yoziladi
df.shape
```

```python
# index bōyicha qatorlarni tartiblash
df.sort_index()
```

```python
# Ustunlari bōyicha tartiblash
df.sort_index(axis=1)
```

```python
# Teskarisiga yani yuqori qiymatdan kichik qiymatga qarab tartiblash
df.sort_index(ascending=False)
```

```python
# Ustunlarga qōllash usuli , teskarisiga yani ohiridan boshiga qarab
df.sort_index(axis=1, ascending=False)
```

```python
# istalgan ustun bōyicha tartiblash
df.sort_values(by='price')
```

```python
# teskarisiga tartiblash
df.sort_values(by='price', ascending=False)
```

```python
# bir nechta ustunlarni tartiblash, narhi va bosib ōtkan km bōyicha
df.sort_values(by=['price', 'mileage'], ascending=False)
```

```python
# ishlab chiqarilgan yili bōyicha tartiblash
df.sort_values(by='year')
```
