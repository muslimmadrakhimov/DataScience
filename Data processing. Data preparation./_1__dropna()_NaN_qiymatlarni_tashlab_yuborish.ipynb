```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

```python
# Mavjud bõlmagan qiymatlarni kõrish usuli, yani mavjud bõlmasa True qaytaradi
df.isnull()
```

```python
# # Mavjud bõlmagan qiymatlarni kõrish usuli, yani mavjud bõlmasa False qaytaradi
df.notnull()
```

```python
# Mavjud bõlmagan qiymatlar sonini chiqarib berish komandasi
df.isnull().sum() # tushib qolgan qiymatlar soni
```

```python
# qatorlar sonini aniqlash
len(df) # 61 ta qator mavjud
```

```python
# proportsiyasini hisoblash
df.isnull().sum()/len(df)*100 # proportsiya hisoblanib 100 ga kõpaytirdik
# masalan wheel-base ustunida 24 % qiymatlar tushib qolgan
```

```python
# NaN qiymatlari bor qatorlarni tashlab yuborish
df.dropna()
```

```python
# qancha qator qolganini aniqlash
len(df.dropna())
```

```python
# NaN qiymati bor ustunlarni tashlab yuborish
df.dropna(axis=1)
```

```python
# Qatorlarni va ustunlarni tashlab yuborishda chegara belgilash, yani qanchadir miqdorda tashlab yuborish
df.dropna(axis=1, thresh=48) # thresh - chegara, kamida 48 ta qiymatli ustunlar qolsin degan komanda
```

```python
# ustundan tanlab NaN qiymati bor qaatorlani tashlab yuborish
df.dropna(subset=['price'])
```

```python
df.dropna(subset=['price', 'horsepower']) # 2 ta ustundagi NaN qatorlarni tashlab yuborish
```
