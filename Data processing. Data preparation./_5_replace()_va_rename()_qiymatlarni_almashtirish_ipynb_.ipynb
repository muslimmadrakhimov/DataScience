```python
import pandas as pd
import numpy as np
```

.map() qiymatlarni moslashtirish(almashtirish)

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

Ustunlarni qiymatlariga õzgartirish kiritish

.map()
.apply()  

```python
# DataFrame da num-of-cylinders ustunida qiymatlar text kõtinishida yozilgan
# shu qiymatlarni son kõrinishiga õzgartiramiz
# qiymatlarini oldin kõrib chiqamiz
df['num-of-cylinders'].unique()
```

```python
# Oldin lugat yaratib olamiz va har biriga sondagi qiymatini berib chiqamiz
dict = {'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'eight':8, 'twelve':12}
df['cylinders'] = df['num-of-cylinders'].map(dict) # yangi 'cylindes' ustun tushdi .map() yordamida
df
```

```python
# price - narh ustunini chiqaramiz
df['price']
```

```python
# price ustunida narhlar dollarda yozilgan
# shu dollarni sõm kõrinishga õtkizamiz
# masalan kurs hozirda 10720 tahminan
kurs = 10720
df['price'].map(lambda x: x*kurs) # lambda bu erda bir martalik funksiya
```

```python
# bu õzgarishlarni 'price_uz' ustuniga saqlab olamiz
kurs = 10720
df['price_uz'] = df['price'].map(lambda x: x*kurs)
df # jadvalni ohida yangi ustun paydo bõldi
```

.replace() va .rename() - qiymatlarni almashtirish

```python
# qiymatlar õrniga qõyib yozish
df.replace('four', 4) # ustunlardagi four qiymatlar õrniga 4 raqam yoziladi
```

```python
# yana bir misol
df.replace('six', 6)
```

```python
# birdaniga bir nechta qiymatlarni õzgartirish
df.replace(dict)
```

```python
# Yangi DataFrame yuklab olamiz
df2 = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population_duplicates.csv', index_col='country')
```

```python
# rename() metodi ustunlarni, qatorlarni nomini õzgartiradi
df2.rename(index=str.upper) # index teks bõlgani uchun str ni qõshib yoziladi
# natija index katta harfga õzgaradi
```

```python
# ustunlarni katta harfga õzgartirish
df2.rename(columns=str.upper)
```

```python
# index va ustunlarni nomlari katta harflar bilan yozish
df2.rename(index=str.upper, columns=str.upper)
```

```python
# ustunlarni ichidan tanlab nomini õzgartirish
# masalan pop2021 ustun nomini population ga õzgartirish
df2.rename(columns={"pop2021":"population"})
# õzgarishlarni saqlab qõyish uchun ohirida inplace=True komandasini beramiz
```
