```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv", index_col=0)
df.head()
```

```python
df.company.unique()
```

```python
# avtolar soni nechtadanligini kõramiz
df.company.value_counts()
```

```python
# hamma kompaniyalarni avtolarni õrtacha qiymatini bilmoqchimiz
# Faqat raqamli ustunlarni tanlash
numeric_cols = df.select_dtypes(include=['number']).columns

# Raqamli ustunlar bo'yicha o'rtacha qiymatni hisoblash
df.groupby('company')[numeric_cols].mean()


```

```python
# eng qimmat mashinalarni õrtacha qiymatini chiqarish
df.groupby('company').max()
```

```python
# Avtolarni kompaniyasi bõyicha guruhlash va narhi bõyicha eng qimmatini chiqarib berish
df.groupby('company').price.max()
```

```python
# Guruhlashda bir nechta ustun bõyicha guruhlash
df.groupby(['company', 'body-style']).price.mean() # õrtacha narhi
```

agg - funksiyasi yordamida bir nechta funksiyalarni kõrsatish mumkin

```python
numeric_cols = df.select_dtypes(include=['number']).columns
df.groupby('company')[numeric_cols].agg(['min', 'max', 'mean'])
```
