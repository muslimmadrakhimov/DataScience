Ma'lumotlarni tōldirish

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

Yõq NaN qiymatlarini õrnini tõldirish

```python
# barcha yōq qiymatlar õrniga standard 0 qiymat berish
df.fillna(0) # fillna tōldirish deb nomlanadi
```

```python
# Ma'lum bir ustunlar uchun qiymat berish
# horsepower ustun uchun 100 qiymat, wheel-base ustun uchun 90 qiymat
df.fillna({'horsepower':100, 'wheel-base':90})
```

```python
# Õrta qiymatlarni olish
np.mean(df['wheel-base']) # wheel-base ustuni uchun õrta qiymatni hisoblash
```

```python
# bu erda horsepower va wheel-base ustunlarida yõq qiymatlar uchun õrta qiymatni kiritish
df.fillna({'horsepower':np.mean(df['horsepower']), 'wheel-base':np.mean(df['wheel-base'])})
```

Mavjud bõlmagan NaN qiymatlarini õrnini tõldirishni yana bir usuli

```python
# Mavjud bõlmagan qiymatlarni õzidan avvalgi qiymat bilan tõldirib ketish
# ffill - oldinga qarab tõldirib ketish
df.fillna(method='ffill')
```

```python
# Mavjud bõlmagan qiymatlarni õzidan keyingi qiymat bilan tõldirib ketish
# bfill - keyingi qarab tõldirib ketish
df.fillna(method='bfill')
```

```python
# Kompaniyasi bõyicha ajratish
df[df.company=='toyota']
```

```python
df[df.company=='toyota'].fillna(method='bfill')
```

```python
# hatchback masinalari uchun õrtacha qiymatni berish
dfh = df[df['body-style']=='hatchback']
```

```python
import pandas as pd

# Применяем fillna только к числовым столбцам
dfh_numeric = dfh.select_dtypes(include=[int, float])

# Заполняем пропущенные значения средними только в числовых столбцах
dfh[dfh_numeric.columns] = dfh_numeric.fillna(dfh_numeric.mean())
dfh
```
