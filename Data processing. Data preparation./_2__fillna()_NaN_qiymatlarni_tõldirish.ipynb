Ma'lumotlarni t≈çldirish

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

Y√µq NaN qiymatlarini √µrnini t√µldirish

```python
# barcha y≈çq qiymatlar √µrniga standard 0 qiymat berish
df.fillna(0) # fillna t≈çldirish deb nomlanadi
```

```python
# Ma'lum bir ustunlar uchun qiymat berish
# horsepower ustun uchun 100 qiymat, wheel-base ustun uchun 90 qiymat
df.fillna({'horsepower':100, 'wheel-base':90})
```

```python
# √ïrta qiymatlarni olish
np.mean(df['wheel-base']) # wheel-base ustuni uchun √µrta qiymatni hisoblash
```

```python
# bu erda horsepower va wheel-base ustunlarida y√µq qiymatlar uchun √µrta qiymatni kiritish
df.fillna({'horsepower':np.mean(df['horsepower']), 'wheel-base':np.mean(df['wheel-base'])})
```

Mavjud b√µlmagan NaN qiymatlarini √µrnini t√µldirishni yana bir usuli

```python
# Mavjud b√µlmagan qiymatlarni √µzidan avvalgi qiymat bilan t√µldirib ketish
# ffill - oldinga qarab t√µldirib ketish
df.fillna(method='ffill')
```

```python
# Mavjud b√µlmagan qiymatlarni √µzidan keyingi qiymat bilan t√µldirib ketish
# bfill - keyingi qarab t√µldirib ketish
df.fillna(method='bfill')
```

```python
# Kompaniyasi b√µyicha ajratish
df[df.company=='toyota']
```

```python
df[df.company=='toyota'].fillna(method='bfill')
```

```python
# hatchback masinalari uchun √µrtacha qiymatni berish
dfh = df[df['body-style']=='hatchback']
```

```python
import pandas as pd

# –ü—Ä–∏–º–µ–Ω—è–µ–º fillna —Ç–æ–ª—å–∫–æ –∫ —á–∏—Å–ª–æ–≤—ã–º —Å—Ç–æ–ª–±—Ü–∞–º
dfh_numeric = dfh.select_dtypes(include=[int, float])

# –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–∏–º–∏ —Ç–æ–ª—å–∫–æ –≤ —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö
dfh[dfh_numeric.columns] = dfh_numeric.fillna(dfh_numeric.mean())
dfh
```
