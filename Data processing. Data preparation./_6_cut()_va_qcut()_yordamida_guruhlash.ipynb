```python
import pandas as pd
import numpy as np
```

.cut() va .qcut() yordamida guruhlash

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population.csv', usecols=['country', 'pop2021', 'area'], index_col='country')
df.head()
```

Davlatlar aholi soni bõyicha turli hil guruhlarga ajratish

```python
df.loc['Uzbekistan']
```

```python
df.loc['India']
```

Keling quyidagi oraliqlarda nechtadan davlat borligini tahlil qilamiz.

1 milliongacha(0-1000)

1-10 mln oraliqda (1000-10000)

10-30 mln oraliqda (10000-30000)

30-50 mln (30000-50000)

50-100 mln (50000-100000)

100-300 mln (100000-300000)

300mln-1.5 mlrd (300000-1500000)

Buning uchun quyidagi bin (nuqtalarni) yaratib olamiz

```python
bins = [0, 1000, 10000, 30000, 50000, 100000, 300000, 1500000]
```

```python
population = df.pop2021
```

```python
population
```

```python
# Guruhlash
guruh = pd.cut(population, bins)
# jadvaldagi qiymatlarni (population) rõyhatdagi (bins) ga bõlamiz
guruh # davlatlarni guruhlarga ajratib berdi
```

```python
# Har bitta guruhda nechtadan davlat borligini kõrish
guruh.value_counts() # masalan: 0 dan 1 mln gacha bõlgan davlatlar soni 72 ..
```

```python
# asl DataFrame ga saqlab olamiz
df['guruh'] = guruh
df.head()
```

Yangi DataFrame chaqirib olamiz va bunda ham guruhlarga bõlishni kõrib chiqamiz

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv", index_col=0)
df.head()
```

```python
# dataframe da narhi(price) bōyicha statistik malumot chiqarish
df.price.describe()
```

```python
# avto larni narhi(price) bõyicha guruhlash
bins = [0, 10000, 20000, 50000]
```

```python
range = pd.cut(df.price, bins)
range
```

```python
range.value_counts()
```

```python
# qiymarlariga nom berish
names = ["arzon", "õrta", "qimmat"]
range = pd.cut(df.price, bins, labels=names)
range.value_counts()
```

```python
# Oraliqlarga bõlish
range = pd.cut(df.price, 3)
range.value_counts()
```

```python
# Oraliqlarga bõlish
range = pd.qcut(df.horsepower, 3)
range.value_counts()
```
